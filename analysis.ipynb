# Streaming Survey Analysis
This notebook loads the dataset, explores variables, and generates key visualizations.

```
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('streaming_survey_dataset.csv')
df.head()
```

## Average Streaming Spend by Income Bracket

```
order = ['<30k','30-50k','50-75k','75-100k','100k+']
df.groupby('IncomeBracket')['TotalStreamingMonthly'].mean().reindex(order).plot(kind='bar')
plt.xlabel('Income Bracket')
plt.ylabel('Average Total Streaming Monthly Spend')
plt.title('Average Streaming Spend by Income Bracket (Reordered)')
plt.tight_layout()
plt.show()
```

## Average Monthly Spending per Service

```
df[['NetflixMonthly','HuluMonthly','DisneyPlusMonthly']].mean().plot(kind='bar')
plt.xlabel('Streaming Service')
plt.ylabel('Average Monthly Spend')
plt.title('Average Monthly Spending per Streaming Service')
plt.tight_layout()
plt.show()
```

## Age vs Streaming Spend

```
plt.scatter(df['Age'], df['TotalStreamingMonthly'])
plt.xlabel('Age')
plt.ylabel('Total Streaming Monthly Spend')
plt.title('Age vs Streaming Spend')
plt.tight_layout()
plt.show()
```

## Cable Cost vs Streaming Spend

```
plt.scatter(df['CableMonthlyCost'], df['TotalStreamingMonthly'])
plt.xlabel('Cable Monthly Cost')
plt.ylabel('Total Streaming Monthly Spend')
plt.title('Cable Cost vs Streaming Spend')
plt.tight_layout()
plt.show()
```
